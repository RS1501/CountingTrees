```python
## Homogeneous Poisson point process

Poisson = PoissonProcess(100)

points = Poisson.generate(W=(25,25))

display(points, title=fr'Homogeneous PPP, $\lambda$ = 100', path="../../img/hppp.eps")

```

```python
## Inhomogeneous Poisson point process

lam = lambda x: np.sin(x[0])*np.cos(x[1])*10

Poisson2 = PoissonProcess(lam)

points = Poisson2.generate(W=(25,25))

display(points, title=fr'Inhomogeneous PPP, $\lambda$(x,y) = 10.sin(x).cos(y) ', window=(25,25), path="../../img/ihppp.eps")

```

```python
## Thomas point process

lam = 0.1
nu = 10
sigma = 0.4

Thomas =ThomasProcess(lam, nu, sigma)

parents, offsprings = Thomas.generate(W=(25,25))

display(parents=parents, offsprings=offsprings, title=fr'Thomas process, $\lambda$ = 0.1, $\nu$ = 10, $\sigma$ = 0.25', window=(25,25), path="../../img/ThomasP.eps")
```

```python
## Thomas point process superposition

lam1 = 0.1
nu1 = 10
sigma1 = 0.4

lam2 = 0.3
nu2 = 7
sigma2 = 0.1

Thomas1 =ThomasProcess(lam1, nu1, sigma1)

Thomas2 =ThomasProcess(lam2, nu2, sigma2)

parents1, offsprings1 = Thomas1.generate(W=(25,25))
parents2, offsprings2 = Thomas2.generate(W=(25,25))


display_superposition(points=[offsprings1, offsprings2],
                      title=fr'Superposition, ($\lambda$,$\nu$,$\sigma$) = (0.1, 10, 0.4), (0.3, 7, 0.1)', window=(25,25),
                      labels=["Thomas Process 1", "Thomas Process 2"], path="../../img/SuperpThomasP.eps")
```
